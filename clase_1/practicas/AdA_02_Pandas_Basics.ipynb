9. Which was the most-ordered item? 

```python
item_orders = chipo.groupby('item_name')['quantity'].sum()
item_orders_sorted = item_orders.sort_values(ascending=False)
print(item_orders_sorted)
most_ordered_item = item_orders.idxmax()
print("El item mas pedido fue: ",most_ordered_item)
```

10. For the most-ordered item, how many items were ordered?

```python
num_items_ordered = item_orders.loc[most_ordered_item]
print ("La cantidad de veces que fue pedido fue: ", num_items_ordered)
```

11. What was the most ordered item in the choice_description column?

```python
choice_des_orders = chipo.groupby('choice_description')['quantity'].sum()
choice_des_orders_sorted = choice_des_orders.sort_values(ascending=False)
print(choice_des_orders_sorted)
most_ordered_choice = choice_des_orders.idxmax()
print("El item mas pedido fue: ",most_ordered_choice)
```

12. How many items were orderd in total?

```python
num_choice_ordered = choice_des_orders.loc[most_ordered_choice]
print ("La cantidad de veces que fue pedido fue: ", num_choice_ordered)
```
